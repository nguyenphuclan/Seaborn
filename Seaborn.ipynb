```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# đọc dataset, cột đầu tiên của tập dữ liệu là cột id
df = pd.read_csv('Pokemon.csv', index_col=0)
df.head()
```

### Dùng scatter plot để so sánh số liệu 'Attack' và 'Defense' của Pokemon.

```python
# scatter plot mặc định
sns.lmplot(x='Attack', y='Defense', data=df)
```

Điều chỉnh các tham số để có được đồ thị như mong muốn.

```python
# `fit_reg=False`: loại bỏ regression line
# `hue='Stage'`: hiển thị màu sắc khác nhau dựa vào 'Stage'
# tham số `hue` giúp thể hiện chiều thứ 3 của thông tin bằng màu sắc
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
```

Đôi khi bạn có thể sử dụng matplotlib để điều chỉnh đồ thị.

Vì các số liệu 'Attack' và 'Defense' là dương nên 2 trục đồ thị nên bắt đầu từ 0.

```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
# tinh chỉnh các trục bằng matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

### Xem số liệu của pokemon dưới dạng box plot.

```python
# Box plot mặc định
sns.boxplot(data=df)
```

Để xem thông số chiến đấu của Pokemon thì ta nên loại bỏ các cột 'Total', 'Stage', 'Legendary'.

```python
# Tạo DataFrame mới không chứa các cột 'Total', 'Stage', 'Legendary'
combat_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
# Box plot mới từ khung dữ liệu mới tạo
sns.boxplot(data=combat_df)
```

Thiết lập theme khác nhau cho đồ thị. Theme mặc định là 'darkgrid'.

```python
sns.set_style('whitegrid')
```

### Violin plot hữu ích hơn khi cho thấy sự phân bố(qua độ dày) thay vì chỉ là số liệu thống kê như box plot.

Sau đây là đồ thị violin thể hiện sự phân bố của 'Attack' đối với hệ chính('Type 1') của Pokemon.

```python
# Violin plot mặc định
sns.violinplot(x='Type 1', y='Attack', data=df)
```

Để điều chỉnh màu thích hợp cho mỗi hệ ta cần tạo ra 1 bảng màu như sau:

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

Bây giờ sử dụng bảng màu ở trên để đổi màu biểu đồ.

```python
# Tham số `palette` để thay đổi bảng màu
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

Violin plot rất lý tưởng để cho thấy sự phân bố. Tuy nhiên, vì ta chỉ có 151 Pokemon trong bộ dữ liệu nên có thể hiển thị từng điểm trên đồ thị mà không sợ dày đặc.

### Swarm plot để hiển thị sự phân bố của mỗi điểm dữ liệu.

```python
# swarm plot mặc định
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

Để kết hợp violin plot và swarm plot ta làm như sau:

* Thiết lập lại kích thước của đồ thị
* Vẽ violin plot với tham số `inner=None` để loại bỏ các thanh bên trong violin
* Vẽ swarm plot với tham số `color='k'`:màu đen và `alpha=0.7`:độ trong suốt

```python
# Thiết lập figure size với matplotlib
plt.figure(figsize=(10,6))
# violin plot
sns.violinplot(x='Type 1', y='Attack', data=df, inner=None, palette=pkmn_type_colors)
# swarm plot
sns.swarmplot(x='Type 1', y='Attack', data=df, color='k', alpha=0.7)
# Đặt tiêu đề cho đồ thị
plt.title('Attack by Type')
```

Để gộp các cột chỉ số của Pokemon thành 1 cột thì ta làm như sau:
* Dùng hàm melt() để gộp
* Giữ các cột mà ta không muốn gộp bằng id_vars=
* Đặt tên cho cột vừa được tạo ra bằng var_name=

```python
# Gộp các cột chỉ số của pokemon
melted_df = pd.melt(combat_df, id_vars=["Name", "Type 1", "Type 2"], var_name="Stat")
melted_df.head()
```

Thực tế số hàng của melted_df gấp 6 lần stats_df.

```python
print(combat_df.shape)
print(melted_df.shape)
```

Vẽ đồ thị mới thể hiện chỉ số của pokemon

```python
plt.figure(figsize=(10,6))
# swarm plot với `dodge=True`: tách các điểm bằng `hue`
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1', dodge=True, palette=pkmn_type_colors)
plt.ylim(0, 260)
# thiết lập vị trí đặt chú thích
plt.legend(bbox_to_anchor=(0, 0), loc=1)
```

### Heatmap plot để thể hiện mối tương quan giữa các cột.

Nếu giá trị dương:2 cột thuận chiều; nếu giá trị âm: 2 cột nghịch chiều; nếu giá trị 0: 2 cột không có mối tương quan nào

```python
# tương quan giữa các cột
correlation = combat_df.corr() * 100
mask = np.zeros_like(correlation)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    # Heatmap `annot=True`: hiện giá trị trong ô, `mask`: chỉ hiện nửa dưới biểu đồ, `cbar=False`: ko hiển thị thanh colorbar
    sns.heatmap(correlation, mask=mask, annot=True, center=0, cbar=False, cmap='RdBu_r')
```

### Factor plot để tách đồ thị theo categorical classes.

```python
# Factor Plot, với `col='Stage'`:chọn categorical class mà mình muốn tách
# `kind='swarm'`: chọn loại đồ thị để vẽ, ở đây là swarm plot
g = sns.factorplot(x='Type 1', y='Attack', data=df, hue='Stage', col='Stage', kind='swarm')
# Xoay x-labels
g.set_xticklabels(rotation=-45)
# Nếu làm như này sẽ chỉ có tác dụng trên plot cuối
# plt.xticks(rotation=-45)
```
